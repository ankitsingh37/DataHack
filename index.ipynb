```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
```

```python
df_f = pd.read_csv("training_set_features.csv")
df_l = pd.read_csv("training_set_labels.csv")
df_test = pd.read_csv("test_set_features.csv")
```

```python
df_xyz = df_l["xyz_vaccine"]
df_seasonal = df_l["seasonal_vaccine"]
submission = pd.read_csv("submission_format.csv")
df_f
```

```python
df_f.info()
```

```python
df_f.head()
```

```python
categories = df_f.columns
categories
```

```python
df_f.isnull().sum()
```

```python
for i in categories:
        cat_mode = df_f[i].mode()[0]
        df_f[i].fillna(cat_mode, inplace=True)

for i in categories:
        cat_mode = df_test[i].mode()[0]
        df_test[i].fillna(cat_mode, inplace=True)


df_f.isnull().sum()
```

```python
df_f.info()
```

```python
for i in categories:
    df_f[i] = df_f[i].astype("category").cat.codes

for i in categories:
    df_test[i] = df_test[i].astype("category").cat.codes
df_f.hhs_geo_region
```

```python
df_test
```

```python
dx_xyz = df_f
dx_seasonal = df_f
```

```python
x_xyz_train, x_xyz_test, y_xyz_train, y_xyz_test = train_test_split(dx_xyz, df_xyz, test_size=0.1, random_state=50)
x_seas_train, x_seas_test, y_seas_train, y_seas_test = train_test_split(dx_seasonal, df_seasonal, test_size=0.1, random_state=50)

model_xyz = RandomForestClassifier()
model_seasonal = RandomForestClassifier()

model_seasonal.fit(x_seas_train, y_seas_train)
```

```python
model_xyz.fit(x_xyz_train, y_xyz_train)
```

```python
predict_xyz = model_xyz.predict_proba(x_xyz_test)[:,1]
predict_xyz
```

```python
predict_seas = model_seasonal.predict_proba(x_seas_test)[:, 1]
predict_seas
```

```python
roc_auc_score(y_seas_test, predict_seas)
```

```python
model_xyz.score(x_xyz_test, y_xyz_test)
```

```python
model_seasonal.score(x_seas_test, y_seas_test)
```

```python
sub_xyz_pred = model_xyz.predict_proba(df_test)[:, 1]
sub_seas_pred = model_seas.predict_proba(df_test)[:, 1]
```

```python
submissions = pd.DataFrame({'respondent_id': submission['respondent_id'],'xyz_vaccine': sub_xyz_pred,'seasonal_vaccine': sub_seas_pred})
```
